```python

import pandas as pd


data = pd.read_csv('../../data/raw/sjia_mike/sJIA_MetaGWAS_MAF0.01.csv')
data.head()
```

```python
# selected columns
# cols = ['Chromosome', 'Position']
kwargs = {
    'hg19' : lambda x: 'chr' + x.Chromosome.astype(str) + ':' + x.Position.astype(str) +
    '-' +  x.Position.astype(str)

}
data = (data
    .assign(**kwargs)
    )
data_out = data.filter(items = ['hg19'])
```

Second liftover done

```python
suc1 = pd.read_table('../../data/raw/sjia_mike/liftover/success.bed',names=['hg38'])
suc2 = pd.read_table('../../data/raw/sjia_mike/liftover/success2.bed',names=['hg38'])
suc1.equals(suc2)
```

```python
suc1.shape
data_out2.head()
```

```python
# join

df = data_out2.join(suc1)
df.head()
```

```python

re1 = {
    0: '#CHR',
    1: 'POS',
    2: 'POS2'
}

chr1 = {
    '#CHR' : lambda x: x['#CHR'].str.lstrip('chr')
}

df_out = (df.hg38.str.split(':|-',expand = True)
                .rename(re1,axis=1)
                .assign(**chr1)
                .join(df, how = 'outer')
)
df_out.head()
```

```python
df_out.shape
```

```python
df_out.eaf.min()
```

```python
kwargs = {
    'beta' : lambda x: np.log(x['OR']),
    'SNP' : lambda x: x['#CHR'].astype(str) + '_' + x['POS'].astype(str) + 
     '_'+ x['other_allele'].astype(str) + '_' + x['reference_allele'].astype(str)
}
col_rename = {
    #'CHR' : '#CHR',
    #'BP' : 'POS',
    'reference_allele':'REF',
    'other_allele':'ALT',
    'beta':'all_inv_var_meta_beta',
    'p.value' : 'all_inv_var_meta_p'

}
keep = ['#CHR','POS',
                'REF','ALT',
                'SNP', 'all_inv_var_meta_p','all_inv_var_meta_beta']
df_out2 = (
    df_out
    .assign(**kwargs)
    .rename(col_rename, axis = 1)
    .filter(items = keep)
)
```

```python
df_out2.shape
df_out2.to_csv('../../data/raw/sjia_mike/summarystats/summary_hg38.txt', sep = '\t',index = False)
```
